## Data driven models

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from statsmodels.tsa.holtwinters import Holt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
plastic.Sales.plot()
```

### Splitting data 

```python
train=plastic.head(40)
test=plastic.tail(20)
```

## Moving average method

```python
plt.figure(figsize=(12,4))
plastic.Sales.plot(label="org")
for i in range(2,30,4):
    plastic['Sales'].rolling(i).mean().plot(label=str(i))
plt.legend(loc='best')
```

## Time series decomposition

```python
decompose_ts_add=seasonal_decompose(plastic.Sales,period=12)
decompose_ts_add.plot()
plt.show()
```

## ACF plots and PACF plots

```python
import statsmodels.graphics.tsaplots as tsa_plots
tsa_plots.plot_acf(plastic.Sales,lags=12)
tsa_plots.plot_pacf(plastic.Sales,lags=12)
```

## Evaluation metric MAPE

```python
def MAPE(pred,org):
    temp=np.abs((pred-org)/org)*100
    return np.mean(temp)
```

## Simple exponential method

```python
ses_model=SimpleExpSmoothing(train['Sales']).fit(smoothing_level=0.2)
pred_ses_model=ses_model.predict(start=test.index[0],end=test.index[-1])
MAPE(pred_ses_model,test.Sales)
```

## Holt's method

```python
hw_model=Holt(train['Sales']).fit(smoothing_level=0.2)
pred_hw_model=hw_model.predict(start=test.index[0],end=test.index[-1])
MAPE(pred_hw_model,test.Sales)
```

## Holts winter exponential smoothing with additive seasonality and additive trend

```python
hw_model_add=ExponentialSmoothing(train['Sales'],seasonal="add",trend="add",seasonal_periods=12).fit()
pred_hw_model_add=hw_model_add.predict(start=test.index[0],end=test.index[-1])
MAPE(pred_hw_model_add,test.Sales)
```

## Holts winter exponential smoothing with multiplicative seasonality and additive trend

```python
hw_model_mul_add=ExponentialSmoothing(train['Sales'],seasonal="mul",trend="add",seasonal_periods=12).fit()
pred_hw_model_mul_add=hw_model_mul_add.predict(start=test.index[0],end=test.index[-1])
MAPE(pred_hw_model_mul_add,test.Sales)
```

## Final model by combining train and test

```python
hw_model_mul_add=ExponentialSmoothing(plastic['Sales'],seasonal="mul",trend="add",seasonal_periods=12).fit()
```

## Model based approach

```python
plastic=pd.read_csv("PlasticSales.csv")
month=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
import numpy as np
p=plastic['Month'][0]
plastic["months"]=0
for i in range(60):
    p=plastic['Month'][i]
    plastic["months"][i]
month_dummies=pd.DataFrame(pd.get_dummies(plastic['months']))
plastic1=pd.concat([plastic,month_dummies],axis=1)
plastic1["t"]=np.arange(1,61)
plastic1["t_squared"]=plastic1["t"]*plastic1["t"]
plastic1.columns
plastic1["log_sales"]=np.log(plastic1["Sales"])
```

## Splitting data 

```python
train=plastic1.head(40)
test=plastic1.tail(20)
```

```python
#linear model

import statsmodels.formula.api as smf

linear_model=smf.ols('Sales~t',data=train).fit()
pred_linear=pd.Series(linear_model.predict(pd.DataFrame(test['t'])))
rmse_linear=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_linear))**2))
rmse_linear
```

```python
#Exponential

exp=smf.ols('log_sales~t',data=train).fit()
pred_exp=pd.Series(exp.predict(pd.DataFrame(test['t'])))
rmse_exp=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_exp))**2))
rmse_exp
```

```python
#quadratic

quad=smf.ols('Sales~t+t_squared',data=train).fit()
pred_quad=pd.Series(quad.predict(test[["t","t_squared"]]))
rmse_quad=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_quad))**2))
rmse_quad
```

```python
#additive seasonality

add_sea=smf.ols('Sales~month',data=train).fit()
pred_add_sea=pd.Series(add_sea.predict(pd.DataFrame(test['month'])))
rmse_add_sea=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

```python
#additive seasonality quadratic

add_sea_quad=smf.ols('Sales~t+t_squared+month',data=train).fit()
pred_add_sea_quad=pd.Series(add_sea_quad.predict(test[["t","t_squared","month"]]))
rmse_add_sea_quad=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad
```

```python
#multiplicative seasonality

mul_sea=smf.ols('log_sales~month',data=train).fit()
pred_mul_sea=pd.Series(mul_sea.predict(pd.DataFrame(test['month'])))
rmse_mul_sea=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_mul_sea))**2))
rmse_mul_sea
```

```python
#multiplicative additive seasonality

mul_add_sea=smf.ols('log_sales~t+month',data=train).fit()
pred_mul_add_sea=pd.Series(mul_add_sea.predict(test[['t','month']]))
rmse_mul_add_sea=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_mul_add_sea))**2))
rmse_mul_add_sea
```

```python
#compare the results

data={"MODEL":pd.Series(["rmse_linear","rmse_exp","rmse_quad","rmse_add_sea","rmse_add_sea_quad","rmse_mul_sea","rmse_mul_add_sea"]),"RMSE_values":pd.Series([rmse_linear,rmse_exp,rmse_quad,rmse_add_sea,rmse_add_sea_quad,rmse_mul_sea,rmse_mul_add_sea])}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(['RMSE_values'])
```
